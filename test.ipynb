```python
import infomap

# Command line flags can be added as a string to Infomap
from IPython.core.display import display
myInfomap = infomap.Infomap("--two-level --directed")

# Access the default network to add links programmatically
network = myInfomap.network()

# Add weight as optional third argument
network.addLink(0, 1, 0.1)
network.addLink(0, 2, 0.11)
network.addLink(0, 3, 0.3)
network.addLink(1, 0, 0.3)
network.addLink(1, 2, 1)
network.addLink(2, 1, 1)
network.addLink(2, 0, 5)
network.addLink(3, 0, 6)
network.addLink(3, 4, 0.6)
network.addLink(3, 5, 6)
network.addLink(4, 3, 6)
network.addLink(4, 5, 6)
network.addLink(5, 4, 6)
network.addLink(5, 3, 6)

# Run the Infomap search algorithm to find optimal modules
myInfomap.run()

print("Found {} modules with codelength: {}".format(myInfomap.numTopModules(), myInfomap.codelength()))

print("Result")
print("\n#node module")
for node in myInfomap.iterTree():
  if node.isLeaf():
    print("{} {}".format(node.physicalId, node.moduleIndex()))
    
```

```python
import community
import networkx as nx
import matplotlib.pyplot as plt

FG = nx.Graph()
FG.add_weighted_edges_from([(1, 2, 0.125), (1, 3, 0.75), (2, 4, 1.2), (3, 4, 0.375)])
for n, nbrs in FG.adj.items():
    for nbr, eattr in nbrs.items():
        wt = eattr['weight']
        if wt < 0.5: print('(%d, %d, %.3f)' % (n, nbr, wt))

# Replace this with your networkx graph loading depending on your format !
G = FG

#first compute the best partition
partition = community.best_partition(G)

#drawing
size = float(len(set(partition.values())))
pos = nx.spring_layout(G)
count = 0.
for com in set(partition.values()) :
    count = count + 1.
    list_nodes = [nodes for nodes in partition.keys()
                                if partition[nodes] == com]
    nx.draw_networkx_nodes(G, pos, list_nodes, node_size = 20,
                                node_color = str(count / size))


nx.draw_networkx_edges(G, pos, alpha=0.5)
plt.show()
```

```python
import json
import pickle
import re

def Merge(dict1, dict2): 
    return(dict2.update(dict1)) 

f = open('./data/test/new_id_ingredients_tokenized_position.json')
new_id_ingredients_tokenized_position = (json.load(f))
f.close()

f = open('./data/test/id_ingredients.json')
id_ingredients = (json.load(f))
f.close()

with open('./vocabulary/absolute_units.pkl', 'rb') as f: # Includes every ingredient present in the dataset.
    units_list1 = pickle.load(f)
    
with open('./vocabulary/relative_units.pkl', 'rb') as f: # Includes every ingredient present in the dataset.
    units_list2 = pickle.load(f)
    
units_list = list(units_list1.keys()) + list(units_list2.keys())
Merge(units_list1, units_list2)
units_list_dict = units_list2

f = open('./data/layer11.json')
original_recipes_data = (json.load(f))[0:50]
f.close()


    
print(units_list)
print(new_id_ingredients_tokenized_position)
print(id_ingredients)

ingrs_quants_units = {}

for key, value in new_id_ingredients_tokenized_position.items():
    
    ingrs_quants_units[key] = []
    
    for value2 in range(0, len(value)):
        
        for i in range(0, len(units_list)):
            
            if (re.search(r"\b" + re.escape(units_list[i]) + r"\b",(id_ingredients[key][value[value2]["id"]])["ingredient"])) is not None:
                
                if re.search(r"[1-9][0-9][0-9]", (id_ingredients[key][value[value2]["id"]])["ingredient"]):
                    
                    ingrs_quants_units[key].append({"ingredient": value[value2]["ingredient"], "quantity": float(re.search(r"[1-9][0-9][0-9]", ((id_ingredients[key][value[value2]["id"]])["ingredient"])).group()),"unit": units_list[i], "quantity (ml)": float(re.search(r"[1-9][0-9][0-9]", ((id_ingredients[key][value[value2]["id"]])["ingredient"])).group()) * int(units_list_dict[units_list[i]])})
                                
                    break
                
                elif re.search(r"[1-9]/[1-9]", (id_ingredients[key][value[value2]["id"]])["ingredient"]):
                    
                    ingrs_quants_units[key].append({"ingredient": value[value2]["ingredient"], "quantity": int(((re.search(r"[1-9]/[1-9]", (id_ingredients[key][value[value2]["id"]])["ingredient"]).group()).split("/"))[0])/int(((re.search(r"[1-9]/[1-9]", (id_ingredients[key][value[value2]["id"]])["ingredient"]).group()).split("/"))[1]), "unit": units_list[i], "quantity (ml)": (int(((re.search(r"[1-9]/[1-9]", (id_ingredients[key][value[value2]["id"]])["ingredient"]).group()).split("/"))[0])/int(((re.search(r"[1-9]/[1-9]", (id_ingredients[key][value[value2]["id"]])["ingredient"]).group()).split("/"))[1])) * int(units_list_dict[units_list[i]])})
                                
                    break
                    
                elif re.search(r"[1-9][0-9]", (id_ingredients[key][value[value2]["id"]])["ingredient"]):
                    
                    ingrs_quants_units[key].append({"ingredient": value[value2]["ingredient"], "quantity": float(re.search(r"[0-9][0-9]", (id_ingredients[key][value[value2]["id"]])["ingredient"]).group()), "unit": units_list[i], "quantity (ml)": float(re.search(r"[0-9][0-9]", (id_ingredients[key][value[value2]["id"]])["ingredient"]).group()) * int(units_list_dict[units_list[i]])})
                                
                    break
                    
                else:
                    
                    ingrs_quants_units[key].append({"ingredient": value[value2]["ingredient"], "quantity": float(units_list_dict[units_list[i]]), "unit": units_list[i], "quantity (ml)": int(units_list_dict[units_list[i]])})
                                
                    break
                
            elif i == len(units_list) - 1:
                
                if re.search(r"[1-9][0-9][0-9]", (id_ingredients[key][value[value2]["id"]])["ingredient"]):
                    
                    ingrs_quants_units[key].append({"ingredient": value[value2]["ingredient"], "quantity": 200, "unit": value[value2]["ingredient"], "quantity (ml)": 200})
                                                
                elif re.search(r"[1-9]/[1-9]", (id_ingredients[key][value[value2]["id"]])["ingredient"]):
                    
                    ingrs_quants_units[key].append({"ingredient": value[value2]["ingredient"], "quantity": int(((re.search(r"[1-9]/[1-9]", (id_ingredients[key][value[value2]["id"]])["ingredient"]).group()).split("/"))[0])/int(((re.search(r"[1-9]/[1-9]", (id_ingredients[key][value[value2]["id"]])["ingredient"]).group()).split("/"))[1]), "unit": value[value2]["ingredient"], "quantity (ml)": int(((re.search(r"[1-9]/[1-9]", (id_ingredients[key][value[value2]["id"]])["ingredient"]).group()).split("/"))[0])/int(((re.search(r"[1-9]/[1-9]", (id_ingredients[key][value[value2]["id"]])["ingredient"]).group()).split("/"))[1]) * 200})
                                                    
                elif re.search(r"[1-9][0-9]", (id_ingredients[key][value[value2]["id"]])["ingredient"]):
                    
                    ingrs_quants_units[key].append({"ingredient": value[value2]["ingredient"], "quantity": 200, "unit": value[value2]["ingredient"], "quantity (ml)": 200})
                                
                else: 
                    
                    ingrs_quants_units[key].append({"ingredient": value[value2]["ingredient"], "quantity": 200, "unit": value[value2]["ingredient"], "quantity (ml)": 200})
       
ingrs_quants_units
```
